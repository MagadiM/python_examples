```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
superstore = pd.read_csv(r"F:\V\a code work place\New folder\superstore ds\Sample - Superstore.csv",encoding='latin-1')
```

```python
# Data info:
print(superstore.head())
```

```python
print(superstore.info())
```

```python
print(superstore.describe())
```

```python
# Clean data
superstore.dropna(inplace=True)
selected_columns = superstore[['Sales', 'Discount', 'Profit']]
print(selected_columns)
```

```python
# Histogram on sales
plt.hist(superstore['Sales'],bins=10)
plt.xlabel('Sales')
plt.ylabel('frequency')
plt.title('distribition of sales')
plt.show()
```

```python
# Bar chart on regions
count_by_region=superstore['Region'].value_counts()
plt.bar(count_by_region.index,count_by_region.values)
plt.xlabel('Region')
plt.ylabel('count')
plt.title('distribution of regions')
plt.show()
```

```python
average_sales_by_state=superstore.groupby('State')['Sales'].mean()
print(average_sales_by_state)
```

```python
average_sales_by_state=superstore.groupby('State')['Sales'].median()
print(average_sales_by_state)
```

```python
average_sales_by_state=superstore.groupby('State')['Sales'].std()
print(average_sales_by_state)
```

```python
#Compute Descriptive Statistics
print("Descriptive Statistics:")
print(superstore.describe())
```

```python
# Visualizing Distributions (Histograms)
superstore.hist(bins=20, figsize=(12, 8))
plt.suptitle("Histograms of Numerical Columns")
plt.show()
```

```python
# Group by 'Category' and calculate the total sales and profit
category_sales_profit = superstore.groupby('Category')[['Sales', 'Profit']].sum()
category_sales_profit
```

```python
# Calculate correlation between 'Sales' and 'Profit'
correlation = superstore[['Sales', 'Profit']].corr()
correlation
```

```python
# Extract the year and month from the 'Order Date'
superstore['Year'] = superstore['Order Date'].dt.year
superstore['Month'] = superstore['Order Date'].dt.month
time_sales_profit = superstore.groupby(['Year', 'Month'])[['Sales', 'Profit']].sum()
time_sales_profit
```

```python
# line chart
superstore['Order Date'] = pd.to_datetime(superstore['Order Date'])
daily_sales = superstore.groupby(superstore['Order Date'].dt.date)['Sales'].sum()
plt.plot(daily_sales.index, daily_sales.values, color='blue', marker='o', linestyle='-', linewidth=2)
plt.title("Daily Sales Over Time", fontsize=16)
plt.xlabel("Order Date", fontsize=12)
plt.ylabel("Sales (USD)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# scatter plot 
sns.scatterplot(data=superstore, x='Sales', y='Profit', color='blue')
plt.title("Scatter Plot of Sales vs Profit", fontsize=16)
plt.xlabel("Sales (USD)", fontsize=12)
plt.ylabel("Profit (USD)", fontsize=12)
sns.scatterplot(data=superstore, x='Sales', y='Profit', hue='Region', palette='deep', style='Category')
plt.show()
```

```python

# Create a boxplot for Sales
plt.figure(figsize=(10, 6))
sns.boxplot(x=superstore['Sales'])
plt.title('Boxplot for Sales')
plt.show()
```

```python
# Create a boxplot for Profit
plt.figure(figsize=(10, 6))
sns.boxplot(x=superstore['Profit'])
plt.title('Boxplot for Profit')
plt.show()
```
